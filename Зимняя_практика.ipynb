# Задание №2

```python
K_value = 1
x0 = 0.5

x_vals = np.linspace(x0 - 0.5, x0 + 0.5, 200)

taylor_linear = fx.series(x, x0, 1).removeO()*x
taylor_quadratic = fx.series(x, x0, 2).removeO()*x

display(taylor_linear)
display(taylor_quadratic)
```

```python
f_num = sym.lambdify(x, fx.subs({K: K_value}), modules=['numpy'])
taylor_lin = sym.lambdify(x, taylor_linear.subs({K: K_value}), modules=['numpy'])
taylor_quad = sym.lambdify(x, taylor_quadratic.subs({K: K_value}), modules=['numpy'])

```

```python
x_vals = np.linspace(x0 - 0.5, x0 + 0.5, 200)
plt.figure(figsize=(10, 6))
plt.plot(x_vals, f_num(x_vals), '--', label='f(x) = -K*sin(3x)*e^(-4x)', color='black',  linewidth=1)
plt.plot(x_vals, taylor_lin(x_vals), ':x', label='Касательная аппроксимация', color='orange',  linewidth=1)
plt.plot(x_vals, taylor_quad(x_vals), '-', label='Квадратичная аппроксимация', color='blue', linewidth=2)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.ylim([-1, 1])
plt.xlim([x0 - 0.5, x0 + 0.5])
```

```python
g = (x**2 + 4*x + 5)/(x + 2)
lim_pnt = - 2
```

```python
display(g.limit(x, lim_pnt))
```

x = -2 - вертикальная асимптота

```python
g_asymptotic = g / x
k = g_asymptotic.limit(x, sym.oo)
b = (g - k * x).limit(x, sym.oo)


asymptote = k * x + b
print(f"Наклонная асимптота: y = {k}x + {b}")
```

```python
dif0 = g
display(Math('f(x) = %s, \\quad \\frac{df}{dx} = %s' %(sym.latex(dif0),sym.latex(sym.diff(dif0)))))
dif1 = sym.diff(dif0)
display(Math('f(x) = %s, \\quad \\frac{df}{dx} = %s' %(sym.latex(dif1),sym.latex(sym.diff(dif1)))))
dif2 = sym.diff(dif1)


critical_points = sym.solve(dif1, x)
print("Точки экстремума:", critical_points)

extrema_types = {}
for point in critical_points:
    second_derivative = dif2.subs(x, point)
    extrema_types[point] = "min" if second_derivative > 0 else "max" if second_derivative < 0 else "saddle"
print("Типы экстремумов:", extrema_types)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D


fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')


for x in np.linspace(1, 0, 15, endpoint=False):

    phi = np.linspace(0, 2 * np.pi, 100)
    z = np.linspace(10 * (1 - x), 10 * (1 - x) + 1, 10)


    phi, z = np.meshgrid(phi, z)


    r = x * np.sin(10 * phi)


    X = r * np.cos(phi)
    Y = r * np.sin(phi)
    Z = z


    ax.plot_surface(X, Y, Z, cmap='Greens')


ax.set_title("3D Ёлочка", fontsize=16, pad=20)
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
ax.set_box_aspect([1, 1, 2])


plt.show()


```
